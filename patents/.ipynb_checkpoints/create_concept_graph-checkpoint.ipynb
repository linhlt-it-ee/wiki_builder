```python
import os
import sys
import random
sys.path.append("../")

import numpy as np
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
from utils import file_util
```

```python
DATA_DIR = "../data/google_patents/us-25000"
NAME_MENTION_DIR = os.path.join(DATA_DIR, "name_mentions")

parent_files = file_util.get_file_name_in_dir_regex(NAME_MENTION_DIR, "name_mention_[0-9]_parents.json")
parent_files.extend(file_util.get_file_name_in_dir_regex(NAME_MENTION_DIR, "name_mention_[0-9][0-9]_parents.json"))

dict_name_mentions = file_util.load_json(os.path.join(DATA_DIR, "dict_name_mentions.json"))
entity_labels = file_util.load_json(os.path.join(DATA_DIR, "entity_labels.json"))
parents = {}
for file_name in parent_files:
    parents.update(file_util.load_json(file_name))
```

```python
name_mention_mapping = {}
for name_mention_id in parents:
    for text in parents[name_mention_id]["name_mention"]:
        if text not in name_mention_mapping:
            name_mention_mapping[text] = set()
        name_mention_mapping[text].add(name_mention_id)
```

```python
def create_concept_graph(doc_id):
    global dict_name_mentions, parent_files, entity_labels
    
    graph = nx.DiGraph()
    for name_mention in dict_name_mentions[doc_id]:
        for name_mention_eid in name_mention_mapping.get(name_mention, []):
            for parent in parents[name_mention_eid]["parents"]:
                path = parent["path"].split(" >> ")
                nx.add_path(graph, path)
    
    print(graph)
    num_nodes_before = len(graph.nodes)
    
    # Graph reduction
    largest_component_nodes = max(nx.connected_components(graph.to_undirected()), key=len)
    graph = graph.subgraph(largest_component_nodes)
    # Prune nodes
    node_central_scores = nx.algorithms.centrality.information_centrality(graph.to_undirected())
    min_central_score = pd.Series(node_central_scores.values()).quantile(.95)
    central_nodes = [node for node, score in node_central_scores.items() if score >= min_central_score]
   
    graph = graph.subgraph(central_nodes)
    num_nodes_after = len(graph.nodes)
    return num_nodes_before, num_nodes_after, graph
```

```python
doc_ids = list(dict_name_mentions.keys())
sample_doc_id = random.sample(doc_ids, 1)[0]
print("Testing with sample:", sample_doc_id)
```

```python
num_nodes_before, num_nodes_after, sample_graph = create_concept_graph(sample_doc_id)
nx.draw(sample_graph, with_labels=True)
nx.write_edgelist(sample_graph, "sample_graph_edgelist.gz")
```

```python
GRAPH_DIR = os.path.join(DATA_DIR, "graphs")
# os.mkdir(GRAPH_DIR)
pbar = tqdm(doc_ids, desc="Creating concept graphs")

for doc_id in pbar:
    graph_path = os.path.join(GRAPH_DIR, f"{doc_id}_graph_edgelist.gz")
    if os.path.exists(graph_path):
        continue
        
    try:
        num_nodes_before, num_nodes_after, graph = create_concept_graph(doc_id)
    except:
        num_nodes_before, num_nodes_after = None, None
        graph = nx.DiGraph()
        
    
    nx.write_edgelist(graph, graph_path)
    pbar.set_postfix({"reduce": f"{num_nodes_before}->{num_nodes_after}"})
```
